```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
volts = np.array([1.1, 1.5, 2.0, 3.1, 4.1])
amps  = np.array([0.37, 0.46, 0.70, 1.15, 1.93])
```

```python
k, m = np.polyfit(volts, amps, deg = 1)
print(k, m)
```

```python
R = 1/k
print(R)
```

```python
x = np.linspace(0, 5)
y = k*x + m
```

```python
%matplotlib widget
plt.plot(volts, amps, 'rx')
plt.axis([0, 5, -0.5, 2.5])
plt.plot(x, y)
plt.show()
```
