```python
from lcapy import *
```

```python
cct = Circuit("""
P1 1 0; down
R1 1 _21 680; right=1.5
R2 _21 _22 6.8e3; right=1.5
C1 _21 _01 47e-9; down=1.5
C2 _22 _02 4.7e-9; down=1.5
W1 _22 2; right=1.0
W2 0 _01;
W3 _01 _02;
W4 _02 _03;
P2 2 _03; down
""")
cct.draw(options='european resistors', scale = 2)
```

```python
H = cct.transfer('P1', 'P2')
```

```python
%matplotlib widget
plt = H.bode_plot((10, 100000))
```
