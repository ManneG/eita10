```python
from lcapy import *
```

```python
cct = Circuit("""
V1 1 0 {u(t)}; down
R1 1 2 100e3; right=1.5
C1 2 _3 1.4e-9; down=1.5
W1 0 _3; right
""")
cct.draw(options='european resistors')
```
